```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Covid Dataset.csv")
data.info()
```

```python
data.nunique()
```

```python
data.drop(columns=['Wearing Masks','Sanitization from Market'],inplace=True)
```

```python
dicto = {"Yes":True,"No":False}
```

```python
datab = data.replace(dicto)
```

```python
datab
```

**DATAB is cleaned lets train**

```python
#import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
#from keras.layers import Dropout
```

```python
X = datab.drop(columns = ['COVID-19'], axis = 1)
Y = datab['COVID-19']

X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=1)
```

```python
#Logistic regression model
model = Sequential()
model.add(Dense(6, input_dim=18, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
X_val.shape
```

```python
model.fit(X_train, y_train, epochs=100, verbose=1,validation_data=(X_test, y_test))
```

```python
score = model.evaluate(X_val, y_val, verbose=0)
model.summary()
print('Validation loss:', score[0])
print('Validation accuracy:', score[1])
```
